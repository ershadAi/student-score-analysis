```python
import numpy as np

names = np.array(["Ali", "Sara", "Omar", "Nilofar", "Ershad"])
subjects = np.array(["Windows", "Word", "Excel", "Access", "PowerPoint"])

scores = np.array([
    [20, 30, 65, 87, 90],
    [67, 87, 56, 43, 98],
    [78, 94, 24, 75, 78],
    [78, 45, 75, 27, 67],
    [67, 85, 93, 100, 75]
])

# Calculate the average score for each student
average_scores = scores.mean(axis=1)

# Find the best student
best_index = np.argmax(average_scores)
best_name = names[best_index]
best_scores = scores[best_index]

print("===== BEST STUDENT =====")
print("Name:", best_name)
print("Average Score:", round(average_scores[best_index], 2))

print("\nSubject Scores:")
for subject, score in zip(subjects, best_scores):
    print(f"{subject}: {score}")
```
